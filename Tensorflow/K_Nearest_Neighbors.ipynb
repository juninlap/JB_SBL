# 생선 분류(KNN)

```python
#도미 데이터 준비
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]

import matplotlib.pyplot as plt #과학계산용 그래프&함수명 사용

plt.scatter(bream_length, bream_weight) #산점도를 그리는 맷플롯립 함수(x,y축 값의로 하는 매개변수 전달)
plt.xlabel('length') #x축 이름 표시
plt.ylabel('weight') #y축 이름 표시
plt.show() # 화면에 출력
```

```python
#빙어 데이터 준비
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
#도미와 빙어의 산점도
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weigth')
plt.show()
```

```python
length = bream_length + smelt_length #리스트 합
weight = bream_weight + bream_weight #리스트 합
```

```python
fish_data = [[l,w] for l, w in zip(length, weight)] #리스트를 2차원 리스트로 변환&생선 데이터 선언
print(fish_data)
```

```python
fish_target = [1] * 35 + [0] * 14 #도미와 빙어의 값을 0,1로 변환&타겟 선언
print(fish_target)
```

```python
#KNN 알고리즘
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
#알고리즘 훈련
kn.fit(fish_data, fish_target)
```

```python
kn.score(fish_data, fish_target) # 정확도 계산
```

```python
kn.predict([[30, 600]]) # 임의의 값 예측
```

```python
print(kn._fit_X)
```

```python
print(kn._y)
```

```python
kn49 = KNeighborsClassifier(n_neighbors=49) #참고 데이터를 49개로한 model

kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)
```
